````python
# imports
import os
import requests
from bs4 import BeautifulSoup
from IPython.display import display , Markdown,update_display
from openai import OpenAI
import json
from dotenv import load_dotenv
import ollama
````

````python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
````

````python
# set up environment
load_dotenv()
openai =OpenAI()
````

````python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
import random
import time
a = int(input())
for chunk in stream:
    reply +=chunk.choices[0].delta.content or ''
    reply = reply.replace("```","").replace("markdown","")
    update_display(Markdown(reply),display_id = display_handle.display_id)
          
"""
````

````python
system_prompt = "You are a helpful technical tutor who answers questions about python code, software engineering, data science and LLMs"
user_prompt = "Please give a detailed explanation to the following question: " + question
````

````python
messages = [
    {"role": "system","content":system_prompt},
    {"role":"user","content":user_prompt}
]
````

````python
# Get gpt-4o-mini to answer, with streaming
stream = openai.chat.completions.create(model=MODEL_GPT,messages=messages,stream = True)
reply =""
display_handle = display(Markdown(""), display_id = True)
for chunk in stream:
    reply+=chunk.choices[0].delta.content or ''
    reply = reply.replace("```","").replace("markdown", "")
    update_display(Markdown(reply),display_id=display_handle.display_id)

````

````python
# Get Llama 3.2 to answer
response = ollama.chat(model=MODEL_LLAMA,messages= messages)
reply = response['message']['content']
display(Markdown(reply))
````

````python
import ollama
from IPython.display import display, Markdown
import time

MODEL_LLAMA = "llama3.2"


# create an empty display cell
display_handle = display(Markdown(""), display_id=True)

# buffer to accumulate chunks
parts = []

# stream=True gives us chunks as they come
for chunk in ollama.chat(model=MODEL_LLAMA, messages=messages, stream=True):
    content = chunk["message"]["content"]
    if not content:
        continue
    parts.append(content)
    
    # update the same notebook cell in-place
    display_handle.update(Markdown("".join(parts)))
    time.sleep(0.02)  # tiny delay to simulate smooth typing

# final result
reply = "".join(parts)
````
